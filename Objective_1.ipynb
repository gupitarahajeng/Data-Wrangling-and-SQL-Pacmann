```python
# import library
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
# read csv data
popular_product_category = pd.read_csv("/content/drive/My Drive/Project_Data/popular_product_category.csv")

# show data
popular_product_category
```

```python
# count missing value
popular_product_category.isna().sum()
```

```python
# count missing value
popular_product_category.isna().sum()/len(popular_product_category)*100
```

```python
# find mode
mod_product_category_name = popular_product_category["product_category_name"].mode()[0]
mod_product_category_name_english = popular_product_category["product_category_name_english"].mode()[0]

# input mode on the missing value
popular_product_category["product_category_name"] = popular_product_category["product_category_name"].fillna(mod_product_category_name)
popular_product_category["product_category_name_english"] = popular_product_category["product_category_name_english"].fillna(mod_product_category_name_english)
```

```python
# check missing value
popular_product_category.isna().sum()
```

```python
# check outlier
popular_product_category.describe()
```

```python
# show data
popular_product_category
```

```python
# check duplicate
popular_product_category.duplicated(keep=False)
```

```python
# remove duplicate
popular_product_category = popular_product_category.drop_duplicates(subset=["order_id"])
```

```python
# show data
popular_product_category
```

```python
# set new index
popular_product_category.set_index("order_id")
```

```python
# pivot data
piv = popular_product_category.pivot_table(values="order_item_id", index=["product_category_name_english"], aggfunc="sum", fill_value=0, margins=False)
```

```python
# sort the data by the top 10 highest order
table_1 = piv.sort_values(by=["order_item_id"], ascending=False).head(10)
table_1
```

```python
# visualize the data
table_plot1 = table_1.plot(kind='bar')
table_plot1.set_title("10 Most Popular Product Category")
table_plot1.set_xlabel("Product Category")
table_plot1.set_ylabel("Total Order")
```
